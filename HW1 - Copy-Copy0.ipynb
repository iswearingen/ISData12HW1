####5. Turn the 311 data into a database. You can use the pandas read_sql command (example [here)]('http://bit.ly/1EoP81v') or the db package from yHat to do this. 

```python
#YOUR CODE HERE
con = sqlite3.connect(":memory:")
df.to_sql('sf_311', con)
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
#YOUR CODE HERE
pd.read_sql("SELECT * from sf_311 WHERE category='Street and Sidewalk Cleaning' And neighborhood='Haight Ashbury'", con)
```

####8. Which type of complaint was most common? 

```python
#YOUR CODE HERE
pd.read_sql('SELECT COUNT(*) as count_cat, category from sf_311 GROUP BY category ORDER BY count_cat desc', con)
```
